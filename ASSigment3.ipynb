```python
def google_citation_by_university()->BeautifulSoup:

    """Get html content of first page of a given university.


    Parameters

    ----------

    org : str

        Organization ID given by Google.


    Returns

    -------

    page_cont : BeautifulSoup object

        Html content of the page.

    """

    url = 'https://www.commonsensemedia.org/app-reviews/disney/user-reviews/adult' 

    reqest = Request(url,headers=headers)

    page = urlopen(reqest)

    page_cont = BeautifulSoup(page, "lxml")

    return page_cont



univ_page= google_citation_by_university()

print(univ_page)
```

```python
div= univ_page.find_all(class_="reveal__content")
rev=[]
for i in div :
    rev+=[i.text]
    print(i.text)
    
```

```python
import pandas as pd
dictionary ={'review':rev}
df=pd.DataFrame(dictionary)
to_csv=df.to_csv('C:\\Users\\bushr\\Desktop\\review.csv')
```
